```python
import pandas as pd
```

```python
lpath = "C:\\Users\\ytsma22c\\Google Drive\\Darby Work\\LOD paper\\LOD_calcs.csv"
lods = pd.read_csv(lpath)
tpath = "C:\\Users\\ytsma22c\\Google Drive\\Darby Work\\LOD paper\\test_calcs.csv"
test = pd.read_csv(tpath)
```

#### Merge by element, instrument, atmosphere, and number range

```python
full = pd.merge(lods, test, how = 'outer',
                on = ['element', 'instrument', 'atmosphere', 'num_range'])
full = full.drop(columns = 'vector')
```

#### Average between number range results

```python
avg = full.groupby(by = ['element', 'instrument', 'atmosphere'], 
                   as_index = False).mean()
stdev = full.groupby(by = ['element', 'instrument', 'atmosphere'], 
                   as_index = False).std()

stdev.columns = ['element', 'instrument', 'atmosphere'] + list(i + '_sd' for i in stdev.columns[3:])

full = pd.merge(avg, stdev, how='outer',on=('element', 'instrument', 'atmosphere'))
display(full)
```

### Export

```python
epath = "C:\\Users\\ytsma22c\\Google Drive\\Darby Work\\LOD paper\\final_results.csv"
full.to_csv(epath, index = False)
```
